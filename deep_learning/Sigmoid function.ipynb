# Sigmoid函数

基本形式   $$ S(x) = \frac{1}{ 1 + e^{-c * x}} $$

值域范围是(0,1)

```python
import matplotlib.pyplot as plt
import math
import numpy as np
```

```python
def sigmoid(x,c):
    return 1 / (1+ np.exp(-c * x))
```

```python
def plot_sigmoid(c):
    x = np.arange( -8,8,0.2)
    y = sigmoid(x,c)
    plt.plot(x,y, label = c)
    plt.legend()
```

```python
c = np.arange(1,10,2)
```

```python
plt.figure(figsize=(20,10))
for i in c :
    plot_sigmoid(i)
    
```

sigmoid函数的形状根据 c值改变， c值越大越陡峭，越接近于步阶函数， c值越小越平滑
